```python
tops3_tourisme = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['tourisme'])).reset_index(drop=True)
```

```python
base_top3 = tops3_tourisme.groupby(['Pays', 'Annee'],as_index=False).mean()
```

```python
base_top3 = base_top3[['Pays','Annee','tourisme']]
```

```python
tops3_politique = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['politique'])).reset_index(drop=True)
tops3_actu = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['actu'])).reset_index(drop=True)
tops3_religion = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['religion'])).reset_index(drop=True)
tops3_statistiques = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['stats'])).reset_index(drop=True)
tops3_reservations = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['reservations'])).reset_index(drop=True)
tops3_guides = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['guides'])).reset_index(drop=True)
tops3_consulat = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['consulat'])).reset_index(drop=True)
tops3_loisirs = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['loisirs'])).reset_index(drop=True)
tops3_sante = base.groupby(['Pays', 'Annee']).apply(lambda x: x.nlargest(3,['sante'])).reset_index(drop=True)
```

```python
tops3_politique = tops3_politique.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_politique = tops3_politique[['Pays','Annee','politique']]
tops3_actu = tops3_actu.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_actu = tops3_actu[['Pays','Annee','actu']]
tops3_religion = tops3_religion.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_religion = tops3_religion[['Pays','Annee','religion']]
tops3_statistiques = tops3_statistiques.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_statistiques = tops3_statistiques[['Pays','Annee','stats']]
tops3_reservations = tops3_reservations.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_reservations = tops3_reservations[['Pays','Annee','reservations']]
tops3_guides = tops3_guides.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_guides = tops3_guides[['Pays','Annee','guides']]
tops3_consulat = tops3_consulat.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_consulat = tops3_consulat[['Pays','Annee','consulat']]
tops3_loisirs = tops3_loisirs.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_loisirs = tops3_loisirs[['Pays','Annee','loisirs']]
tops3_sante = tops3_sante.groupby(['Pays', 'Annee'],as_index=False).mean()
tops3_sante = tops3_sante[['Pays','Annee','sante']]
```

```python
base_top3= base_top3.merge(tops3_politique, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_actu, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_religion, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_statistiques, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_reservations, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_guides, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_consulat, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_loisirs, on = ['Pays', 'Annee'])
base_top3= base_top3.merge(tops3_sante, on = ['Pays', 'Annee'])
```

```python
base_top3.head()
```

```python
base_plate_top3 = base_top3.melt( id_vars=['Pays','Annee'], var_name = [('tourisme', 'politique','actu')], value_name = 'prox_top3')
base_plate_top3.columns=['Pays', 'Annee','Type_contenu','Prox_top3']
```

```python
base_plate_top3.to_csv('base_plate_top3.csv')
```
