```python
df1 = data[['iprspot','ipsuces', 'ipshabt', 'ipudrst','ipgdtim', 'impfun', 'Label_KM']]
df2 = data[['ipfrule', 'imptrad','impsafe','ipbhprp','ipstrgv','ipmodst','Label_KM']]
df3 = data[['iphlppl','impdiff','ipeqopt','ipcrtiv','iplylfr','impenv','impfree','Label_KM']]
df4 = data[['ipadvnt', 'imprich', 'Label_KM']]
df5 = data[['domicil', 'agea', 'gndr', 'formation_num', 'Label_KM']]
```

```python
data['moy_pers'] = data[['ipcrtiv', 'imprich', 'ipeqopt', 'ipshabt', 'impsafe', 'impdiff', 'ipfrule', 'ipudrst', 'ipmodst', 'ipgdtim', 'impfree',
       'iphlppl', 'ipsuces', 'ipstrgv', 'ipadvnt', 'ipbhprp', 'iprspot','iplylfr', 'impenv', 'imptrad', 'impfun']].mean(axis = 1)
```

```python
data['var_pers'] =  data[['ipcrtiv', 'imprich', 'ipeqopt', 'ipshabt', 'impsafe', 'impdiff', 'ipfrule', 'ipudrst', 'ipmodst', 'ipgdtim', 'impfree',
       'iphlppl', 'ipsuces', 'ipstrgv', 'ipadvnt', 'ipbhprp', 'iprspot','iplylfr', 'impenv', 'imptrad', 'impfun']].var(axis = 1)
```

```python
var_pays = data[['cntry','moy_pers','var_pers']].groupby('cntry').mean()
```

```python
var_pays.moy_pers.sort_values()
```

```python
var_age = data[['tranches_age','moy_pers','var_pers']].groupby('tranches_age').mean()
print(var_age.var_pers.sort_values())
print(var_age.moy_pers.sort_values())
```

```python
var_habit = data[['habitat','moy_pers','var_pers']].groupby('habitat').mean()
print(var_habit.var_pers.sort_values())
print(var_habit.moy_pers.sort_values())
```

```python
var_educ = data[['formation','moy_pers','var_pers']].groupby('formation').mean()
print(var_educ.var_pers.sort_values())
print(var_educ.moy_pers.sort_values())
```

```python
var_Label_KM = data[['Label_KM','moy_pers','var_pers']].groupby('Label_KM').mean()
print(var_Label_KM.var_pers.sort_values())
print(var_Label_KM.moy_pers.sort_values())
```
